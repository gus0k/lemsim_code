```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
m, dev = 5, 0.3
```

```python
x = np.linspace(-5, 10, 1000)
plt.plot(x, stats.norm.pdf(x, m, dev))
```

```python
def update_prior(mu, sigma, fixdev, obs):
    n = len(obs)
    avg = np.mean(obs)
    sigma_ = 1 / ((1 / sigma) + (n / fixdev))
    mu_ = ((mu / sigma) + (avg / fixdev)) * sigma_
    return mu_, sigma_
```

## Experiment

1) Fixed std dev is correct

2) Starts with mean far away

```python
realmean = 10
realdev = 1
sigma = 20
mu = 20
```

```python
for i in range(20):
    fig, ax = plt.subplots()
    X = np.linspace(-50, 100, 10000)
    f = stats.norm.pdf(X, realmean, realdev)
    ax.plot(X, f, c='b')
    sample = np.random.normal(realmean, realdev)
    ax.axvline(sample, c='r')
    f2 = stats.norm.pdf(X, mu, sigma)
    ax.plot(X, f2, c='k')
    print(mu, sigma)
    mu, sigma = update_prior(mu, sigma, 2, [sample])
    
```

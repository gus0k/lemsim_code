```python
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import numpy as np
from pathlib import Path
from utils import getsize
```

```python
files = Path('res/')
```

```python
data = {}
for fl in files.glob('*.pkl'):
    with open(fl, 'rb') as fh: data[str(fl)] = pickle.load(fh)
```

```python
KEYS = list(data.keys())
N = len(data[KEYS[0]][1])
L = data[KEYS[0]][1][0]['L']

print(N, L)
```

```python
KEYS = sorted(KEYS)
KEYS
```

```python
fig, ax = plt.subplots(figsize=(20, 16))

for k, (v1, v2) in data.items():
    if 'True' in k:
        print(v1)
        D = np.vstack([v2[n]['history_post_net'] for n in range(N)]).sum(axis=0)
        ax.plot(D, label=k)


for k, (v1, v2) in data.items():
    if 'False' in k:
        print(v1)
        D = np.vstack([v2[n]['history_post_net'] for n in range(N)]).sum(axis=0)
        ax.plot(D, label=k, c='k', linewidth=6)

ax.legend()
```

```python
for k in KEYS:
    d = data[k][1]
    cons_post = np.vstack([d[n]['history_post_net'] for n in range(N)])
    cons_pre = np.vstack([d[n]['history_pre_net'] for n in range(N)])
    fig, ax = plt.subplots(figsize=(16, 12))
    ax.plot(cons_post.sum(axis=0), label='post', c='b')
    ax.plot(cons_pre.sum(axis=0), label='pre', c='k')

    prices = ['r' if p > 12 else 'g' for p in d[0]['allprices'][:, 3]]

    for l in range(L):
        ax.axvspan(l, l+1, facecolor=prices[l], alpha=0.3)
    
    cost = np.sum(d[n]['history_cost'].sum() for n in range(N))
    ax.set_title(k + "Cost {}".format(cost))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data/data.csv', index_col='date')
forcast = pd.read_csv('data/forcast.csv', index_col='date')
data.index = pd.to_datetime(data.index)
forcast.index = pd.to_datetime(forcast.index)
```

```python
col = '277'
d = data[col].values
f = forcast[col].values
```

```python
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111)

start = 7000
end = 8000

ax.plot(d[start:end], label='real', linewidth=3)
ax.plot(f[start:end], label='forcast')
ax.legend()
```
